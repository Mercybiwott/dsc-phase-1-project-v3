# Data Wrangling

## Completeness

### Checking for missing values

```python
# Checking for null values 
df.isna().sum()
```

 Replacing Numerical Columns with the median

```python
#Replacing numerical columns with the median
for i in ["Number.of.Engines","Total.Fatal.Injuries","Total.Serious.Injuries","Total.Minor.Injuries","Total.Uninjured"]:
    df[i].fillna(df[i].median(),inplace= True)
```

 Filling categorical columns

```python
# fill categorical columns with unknown
unknown_columns =['Event.Id', 'Investigation.Type', 'Accident.Number', 'Event.Date', 'Location', 'Country', 'Latitude', 'Longitude', 'Airport.Code',
                   'Airport.Name', 'Injury.Severity', 'Aircraft.damage',  'Aircraft.Category', 'Registration.Number', 'Make', 'Model','Amateur.Built', 
                  'Engine.Type', 'FAR.Description','Schedule','Purpose.of.flight','Air.carrier','Weather.Condition', 'Broad.phase.of.flight', 'Report.Status','Publication.Date']
       
for col in unknown_columns:
    df[col].fillna('Unknown', inplace =True)
```

 All categorical columns are filled with unknown for consistency in data.


```python
# rechecking for missing values
df.isna().sum()
```

### Dropping duplicated rows

## Validity
Ensure only relevant data for the study object remain


```python
# Dropping colums  that are irrelevant to the project objective
df.drop(columns=(['Accident.Number','Location','Country','Latitude','Longitude','Airport.Code','Airport.Name','FAR.Description','Schedule','Air.carrier','Publication.Date']),axis=1, inplace =True)
df.head()
```

#### Cast the date time object to year
This will enable easy comparison of accidents across different years

```python
# Casting date time object to Year
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Year'] = df['Event.Date'].dt.year
df['Month.Abbr'] = df['Event.Date'].dt.strftime('%b')
df['Day.Abbr'] = df['Event.Date'].dt.strftime('%a')
```

```python
# Filtering out accidents that occured before 1980, to focus on recent incidents
df = df[df['Year'] >= 1980]
```

## Visualize Number of Engines 
### Histograms
 The histogram is used to check for the distribution of total number of engines in each aircraft.
 As illustred in the below histogram, aircrafts with one engine were involved in more accidents compared with those with more than one engine
 

```python
# Plotting Histogram for the number of engine
plt.figure(figsize=(12, 8))
df['Number.of.Engines'].plot(kind='hist', bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Number of Engines')
plt.xlabel('Number of Engines')
plt.ylabel('Frequency')
plt.show()
```

## Analysing number of accidents that occur yearly
The analysis has focused on accidents from uear to identify the trends. From the illustration below numberof accidents have reduced over the years.

```python
# Counting the number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()
# Plotting the number of accidents over the years to identify trends
plt.figure(figsize=(10, 6))
accidents_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Number of Accidents Over the Years')
plt.show()
```

## Investigating Injury Severity
An analysis of injury severity provides picture on the extend of injury sustained by individuals in the accidents. As illustrated below, majority of the accidents were not fatal with few total cases.


```python
# Rating injury severity
# Removing blank spaces
df['Injury.Severity'] = df['Injury.Severity'].apply(lambda x: x.split('(')[0].strip())
injury_severity = df['Injury.Severity'].value_counts()
# Plotting injury severity 
plt.figure(figsize=(12, 8))
sns.barplot(x=injury_severity.index, y=injury_severity.values)
plt.xlabel('Injury Severity')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Injury Severity')
plt.show()
```


## Broad Phase of Flight
An analysis of the broad of flight demonstates the distibution of accidents according to the phase of flight. The figure illustrates that most accidents are unknown, however a quite significant number of accidents occur during the landing,take off  and cruise stages.



```python
# Making a list of broad pahse of flight accidents
broad_phase = df['Broad.phase.of.flight'].value_counts()
# Plotting the distribution of accidents according to broad pahse of flight
plt.figure(figsize=(12, 8))
sns.barplot(x=broad_phase.index, y=broad_phase.values)
plt.xlabel('Broad Phase of Flight')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Broad Phase of Flight')
plt.show()
```

### Accident distribution by make
Accidents by Make provides apicture of manufactural technology on incidents of aircraft crashes. Cessna aircrafts were involved in accidents more than any other make.

```python
# Making a list of top 10 aircraft by Makeinvolved in accidents
top_10_Makes = df['Make'].value_counts().head(10)
# Ploting the top 10 makes to understand which aircraft brands have higher accident rates
plt.figure(figsize=(12, 8))
top_10_Makes.plot(kind='bar', color='Blue')
plt.title('Top 10  Makes')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```
